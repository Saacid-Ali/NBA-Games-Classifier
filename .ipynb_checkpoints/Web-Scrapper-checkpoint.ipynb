```python
#Importing Libraries
import requests
from bs4 import BeautifulSoup

l=[]

base_url="https://basketball.realgm.com/nba/allstar/game/rosters/"
for page in range(2004,2021,1):    
    r=requests.get(base_url+str(page))
    c=r.content
    soup=BeautifulSoup(c,"html.parser")
    all=soup.find_all("table", {"class":"basketball compact"})
    names = all[0].find_all("td", {"data-th":"Player"})
    teams = all[0].find_all("td", {"data-th":"Team"})
    selection_types = all[0].find_all("td", {"data-th":"Selection Type"})

    name = all[1].find_all("td", {"data-th":"Player"})
    team = all[1].find_all("td", {"data-th":"Team"})
    selection_type = all[1].find_all("td", {"data-th":"Selection Type"})


    for (a,b,c) in zip(names,teams,selection_types):
        d={}
        d["Name"] = a.text
        d["Team"] = b.text
        d["Selection Type"] = c.text
        l.append(d)
    
    for (e,f,g) in zip(name,team,selection_type):
        h={}
        h["Name"] = e.text
        h["Team"] = f.text
        h["Selection Type"] = g.text
        l.append(h)
    
```

```python
#Transforming the scraped data into a dataframe
import pandas
df=pandas.DataFrame(l)
df
```

```python
#Saving the dataframe to a csv file
df.to_csv("All_Star_Selections.csv")
```
